```python
def step_gradient(points, learning_rate, m , c):
    m_slope = 0
    c_slope = 0
    M = len(points)
    x=points[:,:13]
    y=points[:,13]
    for i in range(M):
        m_slope += (-2/M)* (y[i] -np.dot(m,x[i]) - c)*x[i]
        c_slope += (-2/M)* (y[i] -np.dot(m,x[i]) - c)
    new_m = m - learning_rate*m_slope
    new_c = c - learning_rate*c_slope
    return new_m, new_c
```

```python
def gd(points, learning_rate, num_iterations):
    m = np.zeros(13)
    c = 0
    for i in range(num_iterations):
        m, c = step_gradient(points, learning_rate, m , c)
        print(i, " Cost: ", cost(points, m, c))
    return m, c
```

```python
def cost(points, m, c):
    total_cost = 0
    M = len(points)
    x=points[:,:13]
    y=points[:,13]
    for i in range(M):
        total_cost += (1/M)*((y[i] - np.dot(m,x[i]) - c)**2)
    return total_cost
```

```python
def run():
    data = np.loadtxt("Gradient_Descent_Boston_Dataset_Train.csv", delimiter=",")
    learning_rate = 0.07
    num_iterations = 100
    m, c = gd(data, learning_rate, num_iterations)
    return m, c
```

```python
m,c=run()
```
